```python
documents = [
    {'type': 'passport',
    'number': '2207 876234',
    'name': 'Василий Гупкин'},
    
    {'type': 'invoice',
     'number': '11-2',
     'name': 'Геннадий Покемонов'},
    
    {'type': 'insurance',
     'number': '10006',
     'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def listCommand():
  for doc in documents:
    for key, value in doc.items():
      if key == 'type':
        print (value + " ", end = '')
      else:
        print ('"' + value + '"' + ' ', end = '')
    print()

def peopleCommand():
  docNumber = input('Введите номер документа: ')
  found = False
  for doc in documents:
    if doc['number'] == docNumber:
      print(doc['name'])
      found = True
      break
  if found == False:
    print('Документ не найден')

def shelfCommand():
  docNumber = input('Введите номер документа: ')
  found = False
  for key, value in directories.items():
    for item in value:
      if item == docNumber:
        found = True
        print('Номер полки: ' + key)
        break
    if found == True:
      break
  if found == False:
    print('Документ не найден')
  printDirectories()

def addShelfCommand():
  docNumber = input('Введите номер новой полки: ')
  if docNumber in directories:
    print('Полка уже существует')
  else:
    directories[docNumber] = []
  printDirectories()

def deleteCommand():
  docNumber = input('Введите номер документа: ')
  found = False
  for value in directories.values():
    if docNumber in value:
      value.remove(docNumber)
      found = True
  if found == False:
    print('Документ не найден')
  printDirectories()

def addCommand():
  docType = input('Введите тип документа: ')
  docNumber = input('Введите номер документа: ')
  docOwner = input('Введите имя владельца: ')
  dirNumber = input('Введите номер полки: ')
  if dirNumber in directories:
    newDoc = {
      'type': docType,
      'number': docNumber,
      'name': docOwner
    }
    documents.append(newDoc)
    directory = directories[dirNumber]
    directory.append(docNumber) 
  else:
    print('Полка не найдена')
  printDirectories()

def moveCommand():
  docNumber = input('Введите номер документа: ')
  dirNumber = input('Введите целевую полку: ')
  if dirNumber in directories.keys():
    found = False
    for value in directories.values():
      if docNumber in value:
        value.remove(docNumber)
        directory = directories[dirNumber]
        directory.append(docNumber)
        break
    if found == False:
      print('Документ не найден')
    
  else:
    print('Полки не существует')
  printDirectories()

def getData(command):
  if command == 'list':
    listCommand()
  elif command == 'people':
    peopleCommand()
  elif command == 'shelf':
    shelfCommand()
  elif command == 'as':
    addShelfCommand()
  elif command == 'delete':
    deleteCommand()
  elif command == 'move':
    moveCommand()
  elif command == 'add':
    addCommand()
  else:
    print ("Неопознанная команда")

def printDirectories():
  print('\nПолки:')
  for key, value in directories.items():
    print(key + ':')
    for doc in value:
      print(doc + " ", end = '')
    print()

getData('move')
```
